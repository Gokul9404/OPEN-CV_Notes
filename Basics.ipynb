```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Opening an image

cv2.imread("# path to the image #")

```python
image = cv2.imread("Test_image_2.jpeg")
```

## Showing an image

```python
plt.imshow(image)
plt.show()
```

## <b> Scaling an Image </b>

#### <b>Linear Scaling</b>

```python
image_lin = cv2.resize(image,None,fx=3,fy=3,interpolation=cv2.INTER_LINEAR)
plt.imshow(image_lin)
plt.show()
```

#### <b>Cubic Scaling</b>

```python
image_cubic = cv2.resize(image,None,fx=3,fy=3,interpolation=cv2.INTER_CUBIC)
plt.imshow(image_cubic)
plt.show()
```

#### <b>Arial Scaling</b>

```python
image_are = cv2.resize(image,None,fx=2,fy=2,interpolation=cv2.INTER_AREA)
plt.imshow(image_are)
plt.show()
```

```python
rows = 1
cols = 3
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0].imshow(image_lin)
axes[0].set_title(f"Linear",fontsize = 20)

axes[1].imshow(image_cubic)
axes[1].set_title(f"Cubic",fontsize = 20)

axes[2].imshow(image_are)
axes[2].set_title(f"Ariel",fontsize = 20)

# fig.tight_layout()
plt.show()
```

## <b>Image Translation</b>

```python
# Translation over x and y axis
tx, ty = 50, 10

#Transation matrix
T = np.float32([[1, 0, tx], [0, 1, ty]]) 

# Store height and width of the image
height, width = image.shape[:2]
fin_width = width + tx
fin_height = height + ty
(T)


translated_image = cv2.warpAffine(image, T,(fin_width,fin_height))
# the image using the matrix, T
# img_translation = cv2.warpAffine(image, T, (fin_width, fin_height))


# plt.imshow(img_translation)
plt.imshow(translated_image)
plt.show()
```

## <b>Image Rotation</b>

```python
# Set Angle of Rotation and Scaling Factor
Angle = 45
Scale = 1

# Store height and width of the image
height, width = image.shape[:2]

rotation_mat = cv2.getRotationMatrix2D((width//2,height//2),Angle,Scale)
print(rotation_mat)

rotated_image = cv2.warpAffine(image,rotation_mat, (width,height))

plt.imshow(rotated_image)
plt.show()
```
