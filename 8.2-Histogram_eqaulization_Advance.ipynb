## <b>Advanced Histogram Equalization Method </b>
There are many of ADHE Methods:

* Adaptive Equalization (AHE)

* Contrast Limited Adaptive Equalization (CLAHE)

* Sliding Window Adaptive Histogram Equalization (SWAHE)

### Import libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Load images

```python
image_1 = cv2.imread("Test_image_2.jpeg")
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2GRAY)
```

### <b>Adaptive Equalization</b>
It differs from normal histogram equalization in the respect that the adaptive method enhances the contrast locally.

It divides the image into distinct blocks and computes histogram equalization for each section. 
Thus, AHE computes many histograms, each corresponding to a distinct section of the image. 

~ ADVANTAGE 

* It enhances the local contrast and definitions of edges in all distinct regions of the image.

~ DISADVANTAGE
* AHE overamplifies the noise in relatively homogeneous regions of an image.

### <b>CLAHE Method</b>
Contrast Limited Adaptive Equlization

A further developed version of AHE, that limits the noise generated by histogram equalization by appyling <b>Contrast Limiting Procedure</b>  

Working of this method :-
<i>
* Divide the image into blocks or tiles (88 is common). The, Plot the histogram and check whether to clip or not.
* CDF and transformation function is then computed for each of the blocks. This transformation function is only appropriate for the blocks center pixel.
* All the remaining pixels are transformed wrt. these center pixels using interpolation.
</i>

```python
# Create a CLAHE object (Arguments are optional).
clahe_1 = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
# Apply CLAHE object method
Cl_1 = clahe_1.apply(image_1)

plt.subplot(221),plt.imshow(image_1,cmap='gray');
plt.subplot(222),plt.imshow(Cl_1,cmap='gray');
plt.show()
```
