## Morphological operation
#### It process and adjust a pixel in the image based on it's neighbouring pixels

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(".Images/Test_image_3.png")
K_size = 3 
```

### <b> Erode </b>

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0].imshow(image)
axes[0].set_title(f"Gray Image",fontsize = 20)

Krnl = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
eroded_img = cv2.erode(image,Krnl,iterations=2)
axes[1].imshow(eroded_img)
axes[1].set_title(f"Eroded",fontsize = 20)

# fig.tight_layout()
plt.show()
```

### <b>Dilate </b>

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0].imshow(image)
axes[0].set_title(f"Gray Image",fontsize = 20)

Krnl = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
eroded_img = cv2.dilate(image,Krnl,iterations=2)
axes[1].imshow(eroded_img)
axes[1].set_title(f"Dilated",fontsize = 20)

fig.tight_layout()
plt.show()
```

### <b> Opening </b>

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0].imshow(image)
axes[0].set_title(f"Gray Image",fontsize = 20)

Krnl = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
eroded_img = cv2.morphologyEx(image,cv2.MORPH_OPEN,Krnl)
axes[1].imshow(eroded_img)
axes[1].set_title(f"Opening",fontsize = 20)

fig.tight_layout()
plt.show()
```

### <b> Closing <b/>

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0].imshow(image)
axes[0].set_title(f"Gray Image",fontsize = 20)

Krnl = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
eroded_img = cv2.morphologyEx(image,cv2.MORPH_CLOSE,Krnl)
axes[1].imshow(eroded_img)
axes[1].set_title(f"Opening",fontsize = 20)

fig.tight_layout()
plt.show()
```

<h3> All morphological operation are below </h3>

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

def morphologicalTransformation():
    img = cv.imread('balls2.jpg', 0)
    _, mask = cv.threshold(img, 220, 255, cv.THRESH_BINARY_INV)

    kernel = np.ones((5, 5), np.uint8)

    dilation = cv.dilate(mask, kernel, iterations=2)
    erosion = cv.erode(mask, kernel, iterations=1)
    opening = cv.morphologyEx(mask, cv.MORPH_OPEN, kernel)
    closing = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernel)
    mg = cv.morphologyEx(mask, cv.MORPH_GRADIENT, kernel)
    th = cv.morphologyEx(mask, cv.MORPH_TOPHAT, kernel)

    titles = ["Original Image", "Mask", "Dilation",
              "Erosion", "Opening", "Closing", "Morph_Gradient", "Tophat"]
    images = [img, mask, dilation, erosion, opening, closing, mg, th]

    for i in range(len(images)):
        plt.subplot(2, 4, i+1), plt.imshow(images[i], 'gray')
        plt.title(titles[i])
        plt.xticks([]), plt.yticks([])

    plt.show()

morphologicalTransformation()
```
