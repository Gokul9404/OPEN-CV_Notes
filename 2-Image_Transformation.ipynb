## <b>Image Transformation</b>
Also called <i>Geomateric Transformaion</i>.

Image can be shifted, Rotated & Strechted in multiple ways

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Affine Transformation

```python
img = cv2.imread('.Images/Test_image_1.png')

rows,cols,ch = img.shape
pts1 = np.float32([[0,0],[200,50],[0,200]])
pts2 = np.float32([[10,100],[200,50],[10,250]])

M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

### Perspective Transformation

```python
img = cv2.imread('.Images/Test_image_1.png')

rows,cols,ch = img.shape

pts1 = np.float32([[56,65],[328,52],[28,357],[319,360]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))

img_ = img.copy()
for i in pts1:
    x = int(i[0])
    y = int(i[1])
    cv2.circle(img_,(x,y),2,(0,25,255),4)


plt.subplot(121),plt.imshow(img_),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
