## Contours

<i>
Contours are defined as the line joining all the points along the boundary of an image that are having the same intensity. 

Contours come handy in shape analysis, finding the size of the object of interest, and object detection.
</i>

Steps involved in detecting a contour:
1. Convert Image to Gray-Scale
2. Apply Blurring and then Canny filter (Find edges in the image)
3. Find contour on the edge_image using <i>``findContour()``</i>
3. Draw contour using <i>``drawContour()``</i> on the Main Image

## Import libraries and load image 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

img = cv2.imread('.Images/Contour_and_ConvexHull_use.png')

plt.imshow(img)
plt.show()
```

#### Apply Blur and find edges in the image

```python
# Convert image to Gray-image
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# Apply GaussianBlur
image = cv2.GaussianBlur(img_gray,(3,3),0)

# Find edges in the image
binary_image = cv2.Canny(image,250,cv2.ADAPTIVE_THRESH_MEAN_C)

plt.imshow(binary_image,cmap='gray')
plt.show()
```

### Find contours in the image

```python
method = cv2.RETR_LIST             # OR cv2.RETR_CCOMP  ||OR||  cv2.RETR_EXTERNAL
mode = cv2.CHAIN_APPROX_NONE        # OR cv2.CHAIN_APPROX_SIMPLE

contours, hierarchy = cv2.findContours(binary_image,mode,method)
```

### Draw Contours in the image

```python
Color = (255,0,0)
Line_thickness = 2
Contour_idx = -1        # Contour id in the contour list
```

Method-1

```python
cv2.drawContours(img, contours, Contour_idx, Color, Line_thickness)

img_1 = img.copy()

# Display the final image
plt.imshow(img_1)
plt.show()
```

Method-2

```python
img_2 = img.copy()

# draw each contour seprately 
for contour in contours:
    cv2.drawContours(img_2,contour,Contour_idx, Color, Line_thickness)

plt.imshow(img_2)
plt.show()
```
