# Bluring

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Test_image_2.jpeg")
K_size = 3 

plt.imshow(image)
plt.show()
```

### <b> Identity Kernal </b>

```python
rows = 1
cols = 3
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Kernel of any size
kernel_3 = np.ones((3,3),dtype=np.float16)/9.0
kernel_5 = np.ones((5,5),dtype=np.float16)/25.0
kernel_7 = np.ones((7,7),dtype=np.float16)/49.0

print(kernel_3, "\n", kernel_5, "\n", kernel_7)

img_out_k_3 = cv2.filter2D(img, -1, kernel_3)
axes[0].imshow(img_out_k_3)
axes[0].set_title(f"Blur Image",fontsize = 20)

img_out_k_5 = cv2.filter2D(img, -1, kernel_5)
axes[1].imshow(img_out_k_5)
axes[1].set_title(f"Box Filter",fontsize = 20)

img_out_k_7 = cv2.filter2D(img, -1, kernel_7)
axes[2].imshow(img_out_k_7)
axes[2].set_title(f"Box Filter Norm",fontsize = 20)

fig.tight_layout()
plt.show()
```

### <b> Box Blur / Avg Blur </b>

```python
rows = 1
cols = 3
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

image_blur_fil = cv2.blur(image,(3,3))
axes[0].imshow(image_blur_fil)
axes[0].set_title(f"Blur Image",fontsize = 20)

image_box_fil =  cv2.boxFilter(image, -1, (3,3), normalize=False)
axes[1].imshow(image_box_fil)
axes[1].set_title(f"Box Filter",fontsize = 20)

image_box_fil_norm = cv2.boxFilter(image,-1, (K_size,K_size))
axes[2].imshow(image_box_fil_norm)
axes[2].set_title(f"Box Filter Norm",fontsize = 20)

# fig.tight_layout()
plt.show()
```

### <b> Gaussian Blur </b>

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0].imshow(image)
axes[0].set_title(f"Image",fontsize = 20)

image_gaus_blur = cv2.GaussianBlur(image,(5,5),1)
axes[1].imshow(image_gaus_blur)
axes[1].set_title(f"Gaussian Blur",fontsize = 20)

fig.tight_layout()
plt.show()
```

### <b> Median blur </b>

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0].imshow(image)
axes[0].set_title(f"Image",fontsize = 20)

image_median_blur = cv2.medianBlur(image,5)
axes[1].imshow(image_median_blur)
axes[1].set_title(f"Median Blur",fontsize = 20)

fig.tight_layout()
plt.show()
```

### <b> Billateral Filter </b>

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0].imshow(image)
axes[0].set_title(f"Blur Image",fontsize = 20)

image_bilatteral = cv2.bilateralFilter(image,5,4,4)
axes[1].imshow(image_bilatteral)
axes[1].set_title(f"Bilateral Blur",fontsize = 20)

fig.tight_layout()
plt.show()
```

### <b> Motion blur</b>
Direction based kernel blurring

```python
image = cv2.imread("Test_image_3.png")

k_size = 21
mot_kernel = np.zeros((k_size,k_size))
mot_kernel[int((k_size-1)/2),:] = np.ones(k_size)

motion_kernel = mot_kernel/k_size

img_out_motion_blur = cv2.filter2D(image, -1, motion_kernel)

plt.imshow(img_out_motion_blur)
plt.show()
```
