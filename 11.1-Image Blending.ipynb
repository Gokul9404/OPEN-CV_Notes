## __Image Blending__

_Image blending basically images addition, but different weights are given to images so that it gives a feeling of blending or transparency._

---
__The formula for the image blending is:-__
* __g(x) = (1- $\alpha$) f<sub>&theta;</sub>(x)  +  $\alpha$ f<sub>1;</sub>(x)__

<i>$\alpha$ from 0â†’1</i>

Using the above function, two images are taken to blend together.

---

OpenCV-Python uses the addWeighted() function to blend images. The function and its parameters are as follows.

_dst=cv.addWeighted(src1, alpha, src2, beta, gamma [, dst[, dtype]])_


## Code

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Load image
img_1 = cv2.imread("Test_image_6.png")
img_2 = cv2.imread("Test_image_7.png")

Width, Height = 180, 180

img_1 = cv2.resize(img_1, (Width,Height))
img_2 = cv2.resize(img_2, (Width,Height))

plt.subplot(121),plt.imshow(img_1,cmap='gray')
plt.subplot(122),plt.imshow(img_2,cmap='gray')
plt.subplots_adjust(hspace=0.5,wspace=0.4)
plt.show()
```

```python
# Set the alpha and beta values
alpha = 0.6
beta = 1 - alpha

# Add (or blend) the images
result = cv2.addWeighted(img_1, alpha, img_2,beta, 0)

plt.imshow(result)
plt.show()
```
