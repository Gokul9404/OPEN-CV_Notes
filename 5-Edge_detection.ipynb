```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image =  cv2.imread(".Images/Test_image_1.png")
```

```python
plt.imshow(image)
plt.show()
```

## First order derivative edge filters

### Soble filter

```python
rows=1
cols = 3
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

image_soble_x = cv2.Sobel(image,-1,1,0)
axes[0].imshow(image_soble_x)
axes[0].set_title(f"Soble-X",fontsize = 20)

image_soble_y = cv2.Sobel(image,-1,0,1)
axes[1].imshow(image_soble_y)
axes[1].set_title(f"Soble-Y",fontsize = 20)

image_soble_x_y = cv2.addWeighted(image_soble_x,0.5,image_soble_y,0.5,1)
axes[2].imshow(image_soble_x_y)
axes[2].set_title(f"Soble-X-and-Y",fontsize = 20)


fig.tight_layout()
plt.show()
```

### Scharr filter

```python
rows=1
cols = 3
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

image_scharr_x = cv2.Scharr(image,-1,1,0)
axes[0].imshow(image_scharr_x)
axes[0].set_title(f"Soble-X",fontsize = 20)

image_scharr_y = cv2.Scharr(image,-1,0,1)
axes[1].imshow(image_scharr_y)
axes[1].set_title(f"Soble-Y",fontsize = 20)

image_scharr_x_y = cv2.addWeighted(image_scharr_x,0.5,image_scharr_y,0.5,1)
axes[2].imshow(image_scharr_x_y)
axes[2].set_title(f"Scharr-X-and-Y",fontsize = 20)


fig.tight_layout()
plt.show()
```

## Second Order Derivative edge filters

### Laplacain Filter

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

# Laplacian Filter works on ray image only
# gray_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
image_2 = cv2.imread("Test_image_2.jpeg")
axes[0].imshow(image_2)
axes[0].set_title(f"Gray Image",fontsize = 20)

K_size = 3 
image_leplacian_fil = cv2.Laplacian(image_2,-1,K_size)
axes[1].imshow(image_leplacian_fil)
axes[1].set_title(f"Laplacian Filter",fontsize = 20)

# fig.tight_layout()
plt.show()
```

## Canny Edge detector

The Canny edge detection algorithm is composed of 5 steps:
1. Noise Reduction
2. Gradient Calculation
3. Non-maximum Suppressions
4. Double threshold
5. Edge Tracking by Hysteresis

```python
rows=2
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

axes[0,0].imshow(image)
axes[0,0].set_title(f"Image",fontsize = 10)

image_canny = cv2.Canny(image,85,190)
axes[0,1].imshow(image_canny)
axes[0,1].set_title(f"Canny edge filter",fontsize = 10)

axes[1,0].imshow(image_2)
axes[1,0].set_title(f"Image",fontsize = 10)

image_canny = cv2.Canny(image_2,85,190)
axes[1,1].imshow(image_canny)
axes[1,1].set_title(f"Canny edge filter",fontsize = 10)

fig.tight_layout()
plt.show()
```
