## Histograms

_We can define the histogram of an image as a 2D bar plot. The horizontal axis represents the pixel intensities. The vertical axis denotes the frequency of each intensity._

---
#### Points about Histogram: 
* Histogram of an image provides a global description of the appearance of an image.
* Information obtained from histogram is very large in quality.
* Histogram of an image represents the relative frequency of occurrence of various gray levels in an image

---

#### Histogram Function: 
* _H(r<sub>k</sub>) = n<sub>k</sub>_

n<sub>k</sub> -> freq. of pixel

r<sub>k</sub> -> pixel range

### Import libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Load images

```python
image_1 = cv2.imread("Images/Test_image_4.jpeg")
image_2 = cv2.imread("Images/Test_image_2.jpeg")
```

### Histogram of each channel in an image

```python
b, g, r = cv2.split(image_1)

plt.subplot(221),plt.imshow(image_1), plt.title("Image",fontsize=10)
plt.subplot(222),plt.hist(b.ravel(), 256, [0, 256], color='b'), plt.title("Histogram of Blue-Channel",fontsize=10)
plt.subplot(223),plt.hist(g.ravel(), 256, [0, 256], color='g'), plt.title("Histogram of Green-Channel",fontsize=10)
plt.subplot(224),plt.hist(r.ravel(), 256, [0, 256], color='r'), plt.title("Histogram of Red-Channel",fontsize=10)
plt.subplots_adjust(hspace=0.5)
plt.show()
```

### Histogram of a Gray-image

```python
# Convert COLOR to Gray
img = cv2.cvtColor(image_1, cv2.COLOR_BGR2GRAY)

# Display histogram
plt.subplot(221), plt.hist(img.ravel(), bins=255, rwidth=0.8, range=(0, 255), color='gray'), plt.title("Histogram of Image",fontsize=10)
plt.show()
```
