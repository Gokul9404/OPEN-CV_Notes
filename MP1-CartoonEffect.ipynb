### Import Libraies

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

### _Load the image_

```python
image = cv2.imread("./Example-images/MP_1_example.jpeg")
```

* __First step__ is to convert the image into grayscale. We use the cvtcolor(image, BGR2GRAY) method, this returns the image in grayscale.

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray,cmap='gray')
plt.show()
```

* __Second Step__ is to smoothening of the gray-image.
To smoothen an image, we simply apply a blur effect. This is done using medianBlur() function.

```python
gray_blur = cv2.medianBlur(gray, 3)

plt.imshow(gray_blur,cmap='gray')
plt.show()
```

* __Third Step__ Retrieving the edges of an image

```python
edges = cv2.adaptiveThreshold(gray, 255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)

plt.imshow(edges,cmap='gray')
plt.show()
```

* __Fourth Step__ Preparing a Mask Image

```python
color = cv2.bilateralFilter(image, 9, 280, 250)

plt.imshow(color,cmap='gray')
plt.show()
```

* __Fifth Step__ Giving a Cartoon Effect.
Cartoon effect = Highlighted Edges + Smooth colors

```python
cartoon = cv2.bitwise_and(color, color, mask=edges)

plt.imshow(cartoon)
plt.show()
```

## Step by step transformation in one place

```python
plt.subplot(3, 2, 1)
plt.imshow(image)
plt.subplot(3, 2, 2)
plt.imshow(gray,cmap='gray')
plt.subplot(3, 2, 3)
plt.imshow(gray_blur,cmap='gray')
plt.subplot(3, 2, 4)
plt.imshow(edges,cmap='gray')
plt.subplot(3, 2, 5)
plt.imshow(color)
plt.subplot(3, 2, 6)
plt.imshow(cartoon)
plt.subplots_adjust(wspace=0.1, hspace=0.4)
plt.show()
```
