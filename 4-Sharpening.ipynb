## Sharpening

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(".Images/Test_image_2.jpeg")
K_size = 3 

plt.imshow(image)
plt.show()
```

### <b>Way-1 : Sharpen Kernal </b>

```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))

Kernel_S = np.array([[0,-1,0],
            [-1,5,-1],
            [0,-1,0]
])


axes[0].imshow(image)
axes[0].set_title(f"Blur Image",fontsize = 20)

image_sharp_1  =cv2.filter2D(image,-1,Kernel_S)
axes[1].imshow(image_sharp_1)
axes[1].set_title(f"Sharpend Image",fontsize = 20)

fig.tight_layout()
plt.show()
```

### <b>Way-2 : Reduction Method </b>
Original_image * alpha - Blurred_image * beta => Sharpened Image

aplha - beta always equals to '1'


```python
rows = 1
cols = 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15))


axes[0].imshow(image)
axes[0].set_title(f"Blur Image",fontsize = 20)


Blur_img = cv2.GaussianBlur(image,(5,5),1)
aplha, beta = 2.8, -1.8     # sum of alpha beta should be 1 only
gamma = 0                   # used to increase intensity
image_sharp_2 = cv2.addWeighted(image,aplha,Blur_img,beta,gamma)
axes[1].imshow(image_sharp_1)
axes[1].set_title(f"Sharpend Image",fontsize = 20)

fig.tight_layout()
plt.show()
```
